##### **Q1.** Design a python program which performs the linearregression operation on the given data to predict the house price. Also, visualize the data for different attributes.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
```

```python
data = pd.read_csv("Housing.csv")
df = pd.DataFrame(data)
df
```

```python
plt.title("Heatmap of Housing Data")
sns.heatmap(df.corr(), annot=True)
```

```python
plt.scatter(df["price"],df["area"])
plt.title("Price vs Area")
```

```python
model = linear_model.LinearRegression()
model.fit(df[["area"]],df["price"])
```

```python
model.predict([[3000]])
```

```python
slope = model.coef_
slope
```

```python
intercept = model.intercept_
intercept
```

```python
df["area"]
```

```python
df["price"]
```

```python
plt.plot(df["area"],df["price"],"o")
plt.plot(df["area"],slope*df["area"]+intercept,"r")
plt.title("Price vs Area")
```
