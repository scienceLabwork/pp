##### **Q1.** Develop a python program that reads the image, display matrix representation of an image creates a histogram of the image and apply the smoothing effect on an image.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('p9.jpg')
```

```python
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.show()
```

```python
print(img)
```

```python
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img2, cmap='gray', interpolation='bicubic')
```

```python
img = cv2.GaussianBlur(img, (5,5), 0)
plt.imshow(img, cmap='gray', interpolation='bicubic')
```
