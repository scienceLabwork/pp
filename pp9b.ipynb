##### **Q2.** Develop a python program which takes the video as an argument and extract all the frames from a video. Select specific frames and recreate the video, which has selected frames only.

```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
video = 'sample.mp4'
print("file exists?", os.path.exists(video))
```

```python
cap = cv2.VideoCapture(video)
i = 0
c = 0
framearr = []
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        i+=1
        if(i in [i for i in range(30,91,30)]):
            c+=1
            plt.imshow(frame)
            framearr.append(frame)
            plt.show()
    else:
        break
print("total frames:", c)
print("1 FPS")
```

```python
print("Video Size")
print(framearr[0].shape)
```

# Recreating video with 1st and last frame

```python

convert = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', convert, 1, (framearr[0].shape[1], framearr[0].shape[0]))
out.write(framearr[0])
out.write(framearr[c-1])
out.release()
print("Video Created at output.mp4")
```
